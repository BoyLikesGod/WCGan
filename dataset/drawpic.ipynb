```python
import matplotlib.pyplot as plt
import pickle
```

```python
with open('pheme9_data.pkl', 'rb') as f:
    dataset = pickle.load(f)
with open('pheme9_index.pkl', 'rb') as f:
    index = pickle.load(f)
```

```python
train_index = index['train']
test_index = index['test']
all_index = []
for i in train_index:
    all_index.append(i)
for i in test_index:
    all_index.append(i)
```

```python
charlie_hebdo, ferguson, germanwings_crash, ottawa_shooting, putin_missing, sydney_siege = [], [], [], [], [], []
for i in all_index:
    data_class = dataset[i]['news']
    if data_class == 'charlie-hebdo': charlie_hebdo.append(i)
    elif data_class == 'ferguson': ferguson.append(i)
    elif data_class == 'germanwings-crash': germanwings_crash.append(i)
    elif data_class == 'ottawa-shooting': ottawa_shooting.append(i)
    elif data_class == 'putin-missing': putin_missing.append(i)
    elif data_class == 'sydney-siege': sydney_siege.append(i)
```

```python
print(len(charlie_hebdo), len(ferguson), len(germanwings_crash), len(ottawa_shooting), len(putin_missing), len(sydney_siege))
```

```python
char_num, char_len = [], []
for x, i in enumerate(charlie_hebdo):
    content = dataset[i]['post'].split(' ')
    repost_num = content.count('|') + 1
    word_len = len(content) - repost_num + 1
    char_num.append(repost_num)
    char_len.append(word_len)
print(sum(char_num), sum(char_len))
```

```python
fer_num, fer_len = [], []
for x, i in enumerate(ferguson):
    content = dataset[i]['post'].split(' ')
    repost_num = content.count('|') + 1
    word_len = len(content) - repost_num + 1
    fer_num.append(repost_num)
    fer_len.append(word_len)
print(sum(fer_num), sum(fer_len))
```

```python
ger_num, ger_len = [], []
for x, i in enumerate(germanwings_crash):
    content = dataset[i]['post'].split(' ')
    repost_num = content.count('|') + 1
    if repost_num > 40:
        continue
    word_len = len(content) - repost_num + 1
    ger_num.append(repost_num)
    ger_len.append(word_len)
print(sum(ger_num), sum(ger_len))
```

```python
ott_num, ott_len = [], []
for x, i in enumerate(ottawa_shooting):
    content = dataset[i]['post'].split(' ')
    repost_num = content.count('|') + 1
    if repost_num > 40:
        continue
    word_len = len(content) - repost_num + 1
    ott_num.append(repost_num)
    ott_len.append(word_len)
print(sum(ott_num), sum(ott_len))
```

```python
put_num, put_len = [], []
for x, i in enumerate(putin_missing):
    content = dataset[i]['post'].split(' ')
    repost_num = content.count('|') + 1
    if repost_num > 40:
        continue
    word_len = len(content) - repost_num + 1
    put_num.append(repost_num)
    put_len.append(word_len)
print(sum(put_num), sum(put_len))
```

```python
sy_num, sy_len = [], []
for x, i in enumerate(sydney_siege):
    content = dataset[i]['post'].split(' ')
    repost_num = content.count('|') + 1
    if repost_num > 40:
        continue
    word_len = len(content) - repost_num + 1
    sy_num.append(repost_num)
    sy_len.append(word_len)
print(sum(sy_num), sum(sy_len))
```
